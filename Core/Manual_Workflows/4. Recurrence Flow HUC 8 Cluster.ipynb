```python
import boto3
from datetime import datetime
import json
import os

NWM_DATA_BUCKET = "hydrovis-ti-nwm-us-east-1"
RNR_MAX_FLOWS_DATA_BUCKET = "hydrovis-ti-rnr-us-east-1"
NWM_DATA_BUCKET = "hydrovis-ti-nwm-us-east-1"

def kickoff_viz_lambdas(configurations="all"):
    client = boto3.client('lambda')
    max_flows_lambda = f"viz_max_flows_ti"
    ingest_lambda = f"viz_db_ingest_ti"
    postprocess_lambda = f"viz_db_postprocess_ti"

    all_configurations = [
        "analysis_assim", "analysis_assim_14day", "analysis_assim_hawaii", "analysis_assim_puertorico", "short_range", "short_range_hawaii",
        "short_range_puertorico", "medium_range_mem1", "replace_route"
    ]
    
    conus_configurations = [
        "analysis_assim", "analysis_assim_14day", "short_range", "medium_range_mem1", "replace_route"
    ]
    
    hawaii_configurations = [
        "analysis_assim_hawaii", "short_range_hawaii"
    ]
    
    prvi_configurations = [
        "analysis_assim_puertorico", "short_range_puertorico"
    ]
    
    reference_configurations = [
        "reference"
    ]
    
    if configurations == "all":
        configurations = all_configurations
    elif configurations == "conus":
        configurations = conus_configurations 
    elif configurations == "hawaii":
        configurations = hawaii_configurations 
    elif configurations == "prvi":
        configurations = prvi_configurations
    elif type(configurations) is list:
        invalid_configs = [config for config in configurations if config not in all_configurations]
        if invalid_configs:
            raise Exception(f"{invalid_configs} are not a valid configurations argument")
    elif type(configurations) is str:
        if configurations not in all_configurations and configurations != "reference":
            raise Exception(f"{configurations} are not a valid configurations argument")
        configurations = [configurations]
    else:
        raise Exception(f"{configurations} is not a valid configurations argument")


    for configuration in configurations:
        if configuration == "analysis_assim_14day":
            print("Invoking max flows function for AnA 14 day")
            current_datetime = datetime.utcnow()
            current_date = current_datetime.strftime("%Y%m%d")
            latest_0Z_ana_file = f"common/data/model/com/nwm/prod/nwm.{current_date}/analysis_assim/nwm.t00z.analysis_assim.channel_rt.tm00.conus.nc"  # noqa: E501
            max_flows_payload = {"data_key": latest_0Z_ana_file, "data_bucket": NWM_DATA_BUCKET}

            client.invoke(
                FunctionName=max_flows_lambda,
                InvocationType='Event',
                Payload=bytes(json.dumps(max_flows_payload), "utf-8")
            )
        elif configuration == "reference":
            print(f"Invoking db_postprocess function for reference services")
            postprocess_payload = {
                "configuration": "reference", "reference_time": current_datetime.strftime("%Y-%m-%d %H:%M:%S")
            }
            client.invoke(
                FunctionName=postprocess_lambda,
                InvocationType='Event',
                Payload=bytes(json.dumps(postprocess_payload), "utf-8")
            )
        else:
            print(f"Invoking db_postprocess function for {configuration}")
            postprocess_payload = {"configuration": configuration}
            client.invoke(
                FunctionName=postprocess_lambda,
                InvocationType='Event',
                Payload=bytes(json.dumps(postprocess_payload), "utf-8")
            )
```
